```python
list_final=pd.DataFrame()
extra_f_final=pd.DataFrame()
```

```python
for url in urls:

    list=[]
    extra_f=[]
    browser = webdriver.Chrome(r'C:\Users\lizac\Downloads\chromedriver_win32\chromedriver.exe')
    browser.get(url)
    browser.switch_to.frame('bodyFrame')
    button = browser.find_element_by_xpath("/html/body/blockquote/div/p[10]/font/button[1]")
    button.click()
    html = browser.page_source
    soup = BeautifulSoup(html,'lxml')
    table=soup.find('div',id='valhist')
    rows=table.find_all('tr')
    for row in rows[1:]:
        cols=row.find_all('td')
        cols=[x.text.strip() for x in cols]
        list.append(cols)
    df=pd.DataFrame(list)
    prop_id=soup.find_all('font')[7]
    prop_id = prop_id.get_text(strip=True)
    df['id']=prop_id
    
    
    id=soup.find('div',id="xfsb")
    rows1=id.find_all('tr')
    for row in rows1:
        cols=row.find_all('td')
        cols=[x.text.strip() for x in cols]
        extra_f.append(cols)
    extra_f=pd.DataFrame(extra_f)
    header = extra_f.iloc[0]
    extra_f = extra_f[1:]
    extra_f.rename(columns = header)
    extra_f['id']=prop_id
    list_final = list_final.append(df,ignore_index=True)
    extra_f_final = extra_f_final.append(extra_f,ignore_index=True)
```

```python
list_final.columns=['Year','Homestead Exemption','Just/Market Value','Assessed Value/ SOH Cap','County Taxable Value','School Taxable Value','Municipal Taxable Value','ID']
```
